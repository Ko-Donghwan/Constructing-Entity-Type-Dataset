```python
from collections import defaultdict

# 타입 빈도를 저장할 딕셔너리
type_frequency = defaultdict(int)

# entity2typeid.txt 파일을 열고 각 줄을 읽어서 타입 빈도를 계산
with open('data/entity2typeid.txt', 'r') as file:
    for line in file:
        # 각 줄에서 타입을 추출 (예: "land_reform.n.01 GovernmentOrganization")
        parts = line.strip().split(' ')
        if len(parts) > 1:
            type = parts[1]
            # 타입 빈도를 증가
            type_frequency[type] += 1

# 타입 빈도를 출력
for type, frequency in type_frequency.items():
    print(f'{type}: {frequency}번')
```
