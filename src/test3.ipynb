```python
with open("type2id.txt", "r") as f:
    lines = f.readlines()
```

```python
type_list = []
```

```python
for i in range(0, len(lines)):
    typeName, typeId = lines[i].strip().split(" ")
    type_list.append(typeName)
```

```python
len(type_list)
```

```python
type_dict_IdName = {}
for i in range(0, len(type_list), 1):
    type_dict_IdName[type_list[i]] = i
```

```python
entity_list = []
```

```python
with open('entity2id.txt', 'r') as f:
    lines = f.readlines()

     # 첫 번째 줄(엔티티 수)은 무시하고 두 번째 줄부터 처리
    for i in range(1, len(lines)):
        parts = lines[i].strip().split(' ')  # 공백으로 구분
        if len(parts) == 2:  # '엔티티 번호' 형식일 경우
            entity, index = parts
            entity_list.append(entity)  # 엔티티만 추가
```

```python
len(entity_list)
```

```python
entity_dict_IdName = {}
for i in range(0, len(entity_list), 1):
    entity_dict_IdName[entity_list[i]] = i
```
